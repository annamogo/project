```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df.head()
```

```python
df = df.fillna({"Cabin":0})
df = df.dropna()
df.head()
```

```python
mask = df['Sex'] == "male"
df_m = df[mask]
df_m.head()
```

```python
df_s = df.groupby(by='Sex')
df_f = df_s.get_group("female")
df_m = df_s.get_group("male")
df_f.head()
df_m.head()
```

```python
age = df.groupby('Sex')['Age']
maximum = age.max()
minimum = age.min()
mean = age.mean()
print(maximum)
print(minimum)
print(mean)
```
